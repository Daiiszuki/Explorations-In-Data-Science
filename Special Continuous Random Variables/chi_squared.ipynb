#Chi Squared Distribution


The Chi-Squared distribution is yet another example of a continuous probability distribution. **Primarily used in hypothesis testing**, the shape of the chi-squared distribution is determined by a parameter K (the degrees of freedom)


The graph below demonstrates the effect of different values of K on the chi-squared distribution;

In the real-world, the chi-squared distribution is not a common phenonmenon, thus, it's not used to describe real-world distributions. Instead, it is used mainly in hypothesis tesing. Because of their close relationship to the standard normal distribution, Chi-square distributions are useful in hypothesis testing. Remember that the standard normal distribution is simply one with a mean = 0 and std = 1.



**How exactly the chi-squared distribution related to the normal?**

Imaging we have a standard normal distribution (Z). If we simply square all the values of the sample, we would have a chi squared disribution with K = 1.That is,



> Screenshot 2023-10-13 at 11.05.57 AM.png

Did I hear someone say "Prove it?"

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample from a standard normal distribution (Z)
sample_size = 1000
Z = np.random.normal(0, 1, sample_size)

# Square all the values to get a chi-squared distribution with K = 1
chi_squared = Z**2

# Plot the histograms of both distributions for visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(Z, bins=30, density=True, color='blue', alpha=0.7)
plt.title('Standard Normal Distribution (Z)')
plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.subplot(1, 2, 2)
plt.hist(chi_squared, bins=30, density=True, color='red', alpha=0.7)
plt.title('Chi-Squared Distribution (K = 1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()
```

Now, imaging we take sample from 2 different standard normal distributions (Z1 and Z2). If we repeated the process of sampling a value from each of the normal dists(2 this time), square them and add the result, we would have a chi-squared distribution with K=2.

That is,

More proof?

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples
sample_size = 1000

# Generate samples from two standard normal distributions (Z1 and Z2)
Z1 = np.random.normal(0, 1, sample_size)
Z2 = np.random.normal(0, 1, sample_size)

# Square the values and add them together to get a chi-squared distribution with K = 2
chi_squared = Z1**2 + Z2**2

# Plot the histogram of the chi-squared distribution
plt.figure(figsize=(8, 6))
plt.hist(chi_squared, bins=30, density=True, color='green', alpha=0.7)
plt.title('Chi-Squared Distribution (K = 2)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

**TAKE-AWAY**: If we sampled from K standard normal distributions and square the and sum the values, we will have produced a chi-squared distribution with K degrees of freedom

Notice that the phrase “degrees of freedom” refers to the number of independent standard normal variables involved. The idea is that since these k variables are independent, we can choose them “freely” (i.e., independently).


More formally, the PDF of the chi-square distribution is as follows;

Where;

1. **f(x;k)**: represents the PDF of the chi-squared distribution. Remember that it simply tells us the probability of observing a specific alue of **X**, given a certain number of degrees of freedom **K**


2. **(1/(2 ^ (K/2) * Γ(k/2))**: This is the normalisation constant that ensures that the total area under the chi-squared distribution curve sums up to 1. It accounts for the shape of the distribution and the number of degrees of freedom


3. **x ^ (K/2 - 1)**: represents the "x" raised to the power of (k/ 2-1). This part contributes to the shape of the distribution. As x increases, the probability density decreases, but the rate of decrease depends on "K"

4. **e^(-x/2)**: This term introduces the exponetial decay in the distribution by raising the base of the natural logarithm "e" to the power of (-x/2)


**Summary**: The chi-squared probability density function describes the likelihood of observing a specific value "x" in a chi-squared distribution with "k" degrees of freedom. The shape of the distribution is determined by the term x^(k/2 - 1), and it decreases as "x" deviates from the mean. The exponential term e^(-x/2) contributes to the tail behavior of the distribution. The entire function is normalized to ensure that the total area under the curve sums to 1, making it a valid probability distribution







# Usage


Chi-squared distribution is particularly suited for testing hypotheses related to variances.
It's robust for non-normally distributed data and can be applied to a wide range of real-world problems. It's a fundamental distribution in statistical analysis, providing a theoretical basis for hypothesis testing.Let's use an exaggerated example to illustrate this.

Imagine you are conducting a quality control inspection in a factory that produces light bulbs. You want to test the variance of the lifespans of these bulbs to ensure they meet quality standards.

**Normal Distribution:** If the lifespans of the bulbs are normally distributed, you can use the normal distribution to analyze the data. In this case, let's assume that you expect the lifespans to follow a normal distribution, which means the data is symmetric and bell-shaped.

**Chi-Squared Distribution:** Now, let's consider an exaggerated scenario. Suppose there is a quality issue in the factory, resulting in a situation where some bulbs have extremely short lifespans (outliers) while others have very long lifespans. This creates a highly skewed distribution with a significant number of outliers.

In this exaggerated example, you might encounter cases like:

90% of the bulbs have a typical lifespan.
5% of the bulbs burn out extremely quickly (outliers).
5% of the bulbs last much longer than usual (outliers).
Why Chi-Squared Over Normal?

Using the normal distribution in this case might not be appropriate because it assumes that data is symmetric and does not handle extreme outliers well.
The chi-squared distribution, on the other hand, is more robust in the presence of outliers and is particularly useful for testing variances. By squaring values and summing them, it amplifies the impact of outliers, making them more visible in the statistical analysis. Some example use cases of the chi-squared distribution include;

1. **Chi-Squared Test of Independence**: It's used to determine whether there's a significant association between two categorical variables in a contingency table.

2. **Chi-Squared Goodness of Fit Test**: It assesses how well an observed distribution fits an expected distribution.

3. **Chi-Squared Test for Variance:** It's used to test whether the variance of a population is equal to a specific value.

4. **Confidence Intervals for Variance:** Chi-squared distributions are used to construct confidence intervals for population variances.

#Appendix




1. **Chi-Squared Distribution:** A probability distribution that describes the sum of the squares of k independent standard normal random variables. It is often denoted as χ² (chi-squared).

2. **Degrees of Freedom (k):** A parameter in the chi-squared distribution that determines the shape of the distribution. The number of degrees of freedom, denoted as "k," influences the spread and peakedness of the distribution.

3. **Probability Density Function (PDF):** The mathematical function that defines the chi-squared distribution. It describes the likelihood of observing a specific value "x" for a given number of degrees of freedom "k."

4. **Normalization Constant:** A constant in the chi-squared PDF that ensures the area under the distribution curve sums to 1. It accounts for the shape of the distribution and the number of degrees of freedom.

5. **Chi-Squared Test:** A statistical test that uses the chi-squared distribution to assess the relationship between categorical variables, goodness of fit, or the variance of a population.

6. **Chi-Squared Test of Independence:** A chi-squared test used to determine whether two categorical variables are independent or if there is a significant association between them.

7. **Chi-Squared Goodness of Fit Test:** A statistical test that evaluates how well the observed distribution of data matches an expected (theoretical) distribution. It's used for testing hypotheses about data fit.

8. **Chi-Squared Test for Variance:** A chi-squared test used to compare the sample variance to a hypothesized population variance. It assesses whether the variance is equal to a specific value.

9. **Confidence Intervals for Variance:** Chi-squared distributions are used to construct confidence intervals for population variances. These intervals provide a range within which the true population variance is likely to lie.

10. **Normal Distribution:** A symmetric probability distribution that is often used as a baseline in statistics. The chi-squared distribution can be used when data deviates significantly from a normal distribution.

11. **Outliers:** Data points that significantly deviate from the majority of the data. The chi-squared distribution is robust in the presence of outliers, making it suitable for analyzing such data.

12. **Gamma Function (Γ):** A complex-valued function that extends the concept of a factorial to real and complex numbers. It is used in the chi-squared PDF to calculate the normalization constant.
