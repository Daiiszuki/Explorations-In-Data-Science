# Student's T-Distribution

**What is the t-distribution?**

The t-distribution characterises how sample means, in situations where the standard deviation is unknown, deviate from from the population mean. This is especially relevant when the data originates from a normally distributed.

From a visual perspective, the t-distribution closely resembles a normal distribution but typicall has thicker tails.These thicker tails, as you may be aware, accomodatate greater variability in variables, reflecting increased uncertainity. That is, wider tails mean that there's more room for things to be different from average, wich makes things less certain.

The one-sample t-test is used to compare the sample mean to a known or hypothesized population mean. The formula for the t-score in a one-sample t-test is as follows:

> t = (X - ) / (s / n)

t: The t-score.

X: The sample mean.

: The population mean you are testing against (hypothesized mean).

s: The sample standard deviation.

n: The sample size.



In this context, the t-score measures how many standard errors the sample mean is away from the hypothesized population mean. A larger t-score indicates a greater difference between the sample mean and the population mean. Alternatively, we may interested in calculating the 2-sample t-score, in regard to the '2sample-t-test'


The **two-sample t-test** is used to compare the means of two independent groups to determine if there is a statistically significant difference between them. The formula for the t-score in a two-sample t-test is as follows:

> t = (X - X) / ((s/n) + (s/n))

t: The t-score.

X and X: The sample means of the two groups being compared.

s and s: The sample standard deviations of the two groups.

n and n: The sample sizes of the two groups.

In this context, the t-score quantifies the difference in means between the two groups relative to the variability within each group. A larger t-score indicates a more significant difference between the group means.


**Interpreting the T-Score:**

To interpret the t-score, you typically compare it to a critical value from the t-distribution (based on the degrees of freedom and your desired level of significance) or use it to calculate a p-value. If the t-score is significantly different from zero (or the critical value), you may reject the null hypothesis (in favor of the alternative hypothesis) and conclude that there is a statistically significant difference or effect.


**Whats the key differerence between the T and Normal (Z) Distribution?**

The z-distribution assumes that the population standard deviation is already known, whereas the  the T-distribution relies on the sample standard deviation. The T-distribution is thus, commonly used for smaller sample sizes where the variance is unknown. The main differences and similarities between the normal and T-distribution are oulined below;

- Just like the normal distribution, the t-distribution had has a symmetrical bell shape; Folding it in half at the mean, both sides of the t-distribution look the same, much like the normal distribution.

- Similar to the standard normal distribution, the t-distribution had a mean of 0

- However, the distinction between z and t distributions lie in the fact that the normal distribution assumes knowledge of the population standard deviation.

- Instead, the T-distribution is defined by something called the 'degrees of freedom,' which relates to the sample size (n)

- The t-distribution is thus more handy in cases where we have a small sample size or when there is uncertainity about the population standard deviation, or even both.

- As the sample size increases, the t-distribution starts resembling the normal distribution






**Degrees Of Freedom**

Degrees of Freedom (DF) in the t-distribution represent the number of values in the final calculation of the t-statistic that are free to vary. In other words, they indicate how much uncertainty is involved in your sample data.

Here's why degrees of freedom matter:

- When you take a sample from a population, you use one degree of freedom to calculate the sample mean. This is because, once you've calculated the mean, you've used up one piece of information, and the rest of your data must "agree" with that mean.

- If you knew the population standard deviation (which you often don't), you'd use another degree of freedom to calculate the standard error of the sample.

- So, when you calculate the t-statistic, you have one less degree of freedom than the number of data points in your sample. This is why you typically see 'n-1' degrees of freedom when working with the t-distribution, where 'n' is the number of data points in your sample.

In simpler terms, degrees of freedom in the t-distribution reflect how much room there is for variability in your sample data. They're particularly important when working with small sample sizes because they account for the additional uncertainty introduced by using sample statistics instead of population parameters.

The PDF of the Student's t-distribution, denoted as f(t; n), describes the probability of observing a specific value t for a random variable t that follows this distribution. The elements of the PDF are as follows:

1. **t**: This is the variable you're interested in, and you want to find the probability density at a specific value of t.

2. **n**: The parameter "n" represents the degrees of freedom of the t-distribution. Degrees of freedom determine the shape of the distribution. In the context of the t-distribution, degrees of freedom are typically associated with the size of your sample. If you have a larger sample, you have more degrees of freedom, which makes the t-distribution approach the standard normal distribution.

Now, let's break down the formula for the PDF:

**f(t; n) = (((n+1)/2) / [(n) * (n/2)]) * (1 + (t^2/n))^(-(n+1)/2)**

Here's what each part of the formula represents:

- **((n+1)/2)**: This is the gamma function applied to (n+1)/2. The gamma function is a generalization of the factorial function for non-integer values (allows you to calculate the factorial of non-integer or real numbers). It's used to ensure that the t-distribution is properly normalized.

- **(n/2)**: This is the gamma function applied to n/2. Again, it's part of the normalization process.

- **n**: The degrees of freedom parameter, as mentioned earlier.

- ****: Represents the mathematical constant Pi.

- **t**: The value at which you want to find the probability density.

- **(1 + (t^2/n))^(-(n+1)/2)**: This part of the formula is often referred to as the kernel. It's a function of t and n and determines the shape of the t-distribution. Essentially, it describes how the density of the distribution changes as you move away from the mean (which is 0 in the case of the standard t-distribution).

In summary, the PDF of the t-distribution tells you how likely it is to observe a specific value "t" for a random variable following this distribution, based on the degrees of freedom "n." The shape of the distribution depends on "n," and as "n" increases, the t-distribution approaches the standard normal distribution.

```python
np.random.seed(1)
N = 1000000
df = 10
t_data = np.random.standard_t(df, N)

sns.histplot(t_data, color='green', alpha=0.8, stat='density', bins=100)
sns.kdeplot(t_data, color='black')
plt.xlim(-4,4)
plt.title(f"Student's T Distribution (df={df})");
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy.stats import t

# Set the random seed for reproducibility
np.random.seed(1)

# Number of data points
N = 100000

# Generate x values
x = np.linspace(-5, 5, N)

# Degrees of freedom values
df1, df2, df3, df4, df5, df6 = [2, 3, 5, 10, 20, 100]

# PDF values for each degree of freedom
t_data1 = t.pdf(x, df1)
t_data2 = t.pdf(x, df2)
t_data3 = t.pdf(x, df3)
t_data4 = t.pdf(x, df4)
t_data5 = t.pdf(x, df5)
t_data6 = t.pdf(x, df6)

# Neon-like colors
neon_colors = ['#39FF14', '#FF00FF', '#00FFFF', '#FF1493', '#00FF00', '#FF4500']

# Create a black background
plt.figure(figsize=(10, 6))
plt.gca().set_facecolor('black')

# Plot the lines
plt.plot(x, t_data1, color=neon_colors[0])
plt.plot(x, t_data2, color=neon_colors[1])
plt.plot(x, t_data3, color=neon_colors[2])
plt.plot(x, t_data4, color=neon_colors[3])
plt.plot(x, t_data5, color=neon_colors[4])
plt.plot(x, t_data6, color=neon_colors[5])

# Create legend patches
legend_patches = [mpatches.Patch(color=color, label=f'df={df}') for color, df in zip(neon_colors, [df1, df2, df3, df4, df5, df6])]


plt.legend(handles=legend_patches, frameon=True)

# Set plot limits and add a title
plt.xlim(-5, 5)
plt.ylim(bottom=0)
plt.title('T-student Distribution with different dfs')

# Display the plot
plt.show()
```

We can find a point that has a right probability of
p using the code below:

Additionally, we can find a point that has a left probability of p using;

` t.isf(1-p, df)`

```python
p = 0.95
df = 2
print(f'The point in the T-student(n={df}) Distribution that has a right probability of *{p}*: ', t.isf(p, df))
print(f'The point in the T-student(n={df}) Distribution that has a left probability of *{p}*: ', t.isf(1-p, df))
```

To find the left probability of a point use the code below:

 `t.cdf(X,df)`

To find the right probability of a point use the code below:

`t.sf(X,df)`



```python
X = 0.816497
df = 2
print(f'The left probability of *{X}* in the T-student(n={df}) Distribution is: ', t.cdf(X,df))
print(f'The Right probability of *{X}* in the T-student(n={df}) Distribution is: ', t.sf(X,df))
```

```python
fig, ax = plt.subplots()
#Distribution curve
x= np.arange(-4, 4, 0.001)
ax.plot(x, t.pdf(x, df))
ax.set_title(f'T-student Dist. (n={df})')
ax.set_xlabel('X-Values')
ax.set_ylabel('PDF(X)')

#Fill_between
px=np.arange(X, 4, 0.001)
ax.set_ylim(0, 0.4)
ax.set_xlim(-4, 4)
ax.fill_between(px, t.pdf(px, df), color='#DFFF00')

prob = t.sf(X, df)
ax.text(-1.5, 0.02, f'P(X>{np.round(X, 3)})\n {np.round(prob, 2)}', fontsize=18);
```

#Usage



The t-distribution is used in statistical analysis under specific circumstances, particularly when dealing with sample data, and it's preferred over other distributions in the following situations:

1. **Small Sample Sizes**: The t-distribution is especially useful when working with small sample sizes (typically when n < 30). In these cases, the population standard deviation is often unknown, and the sample standard deviation is used to estimate it. The t-distribution accounts for the extra uncertainty introduced by using the sample standard deviation.

2. **Unknown Population Standard Deviation**: When you don't know the population standard deviation, you can use the t-distribution for inference. The z-distribution, or standard normal distribution, requires knowledge of the population standard deviation, which is rarely available when working with sample data.

3. **Estimating Population Parameters**: The t-distribution is commonly used in the construction of confidence intervals and hypothesis testing for population means. When working with sample means, the t-distribution provides a more accurate estimate of the population mean.

4. **Reduced Assumptions**: The t-distribution is robust against deviations from normality in the data. It's more forgiving than the normal distribution and can be used when the assumption of normality is questionable, especially for small samples.

5. **Accounting for Degrees of Freedom**: The t-distribution is parameterized by degrees of freedom (df). As the sample size increases, the t-distribution approaches the standard normal distribution. This makes the t-distribution versatile for varying sample sizes, and you can use it in a wide range of scenarios.

6. **Heteroscedastic Data**: In some cases, the variance in different groups or subpopulations may not be the same. The t-distribution can still be used, provided that the assumption of independence within the groups holds, and each group's data follows a t-distribution.

In contrast, the standard normal (z) distribution is used when you know the population standard deviation, have a large sample size, and are primarily concerned with the distribution of the sample mean. Other distributions, such as the chi-squared, F, and Poisson distributions, are used in different statistical contexts based on the nature of the data and the specific hypothesis or parameter being tested.

# Appendix


**Definition:**
The Student's t-distribution, also known as the t-distribution, is a probability distribution used in statistical analysis. It's a symmetric probability distribution characterized by its degrees of freedom parameter (df). The t-distribution is related to the normal (z) distribution but is more appropriate when dealing with small sample sizes or when the population standard deviation is unknown.

**Parameters:**
The primary parameter of the t-distribution is the degrees of freedom (df), which determines the shape of the distribution. The higher the degrees of freedom, the closer the t-distribution resembles the standard normal distribution.

**Probability Density Function (PDF):**
The probability density function of the t-distribution is given by:

f(t; df) = (((df+1)/2) / [(df) * (df/2)]) * (1 + (t^2/df))^(-(df+1)/2)

Where:
- f(t; df) is the probability density at a specific value t for given degrees of freedom (df).
-  is the gamma function.
-  is the mathematical constant Pi.
- t is the random variable.
- df is the degrees of freedom.

**Properties:**
1. **Symmetry**: The t-distribution is symmetric around the mean (usually 0) and has heavier tails compared to the standard normal distribution.

2. **Degrees of Freedom**: The degrees of freedom parameter, df, influences the shape of the distribution. As df increases, the t-distribution approaches the standard normal distribution.

3. **Confidence Intervals**: The t-distribution is commonly used to calculate confidence intervals for population means when the population standard deviation is unknown.

4. **Hypothesis Testing**: In hypothesis testing, the t-distribution is used to compare sample means and make inferences about population means.

5. **Robustness**: The t-distribution is robust against deviations from normality, making it suitable for small sample sizes or when the assumption of a normal distribution is questionable.

6. **Convergence to Normal Distribution**: As the degrees of freedom increase, the t-distribution converges to the standard normal distribution, allowing it to be used for larger sample sizes.

**Use Cases:**
The t-distribution is commonly used in various statistical scenarios, including:
- Confidence interval estimation for population means.
- Hypothesis testing when the population standard deviation is unknown.
- Analysis of small sample sizes.
- Quality control in manufacturing and process improvement.
- Student's t-test for comparing two sample means.

**Examples:**
1. If you have a sample of 10 measurements and want to estimate the mean of the population, you would use the t-distribution with 9 degrees of freedom for your confidence intervals.

2. In a scientific study, when comparing the means of two groups with small sample sizes, a t-test is often employed using the t-distribution.

3. In financial risk analysis, the t-distribution is used to model the uncertainty of asset returns, especially when historical data is limited.

**Notable Facts:**
- The t-distribution was introduced by William Sealy Gosset in 1908 and is often referred to as "Student's t" due to Gosset's pen name "Student."

- The t-distribution is a key element in the development of statistical theory and is essential in many areas of research and industry, such as economics, biology, and quality control.

- The t-distribution is included in standard statistical software packages and calculators, making it readily accessible for data analysis.
